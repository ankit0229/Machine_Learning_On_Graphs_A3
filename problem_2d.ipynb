```python
import networkx
import math
import random
import copy
import matplotlib.pyplot as plt
from collections import Counter
from tqdm import tqdm 
import pickle
import sys 
```

```python
Picklefile1 = open('dictionary_betweenness_algo1', 'rb')
dict_edge_betweenness_algo1 = pickle.load(Picklefile1)

Picklefile2 = open('dictionary_betweenness_algo2', 'rb')
dict_edge_betweenness_algo2 = pickle.load(Picklefile2)

combined_list_1 = [[e, dict_edge_betweenness_algo1[e]] for e in dict_edge_betweenness_algo1]
combined_list_2 = [[ed, dict_edge_betweenness_algo2[ed]] for ed in dict_edge_betweenness_algo2]

combined_list_1.sort(key= lambda x:x[1], reverse=True)
combined_list_2.sort(key= lambda x:x[1], reverse=True)

x_cord = [i for i in range(1, 3985)]
y_cord_1 = [math.log10(j[1]) for j in combined_list_1]
y_cord_2 = [math.log10(k[1]) for k in combined_list_2]

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(9,9))
plt.plot(x_cord, y_cord_1, label = "For Algo 1")
plt.plot(x_cord, y_cord_2, label = "For Algo 2")
plt.xlabel('Rank', fontsize = 18)
plt.ylabel('log centrality value', fontsize = 18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend() 
plt.show()
```
